```python
import pandas as pd
import numpy as np

df = pd.read_csv("Downloads/myexcel.csv")  # or use pd.read_excel() if it's an Excel file
df['height'] = np.random.randint(150, 181, size=len(df))
```

```python
df.head()
```

```python
team_counts = df['Team'].value_counts()
team_percentages = df['Team'].value_counts(normalize=True) * 100

import matplotlib.pyplot as plt
team_counts.plot(kind='bar', title='Employee Distribution by Team')
plt.ylabel('Number of Employees')
plt.show()
```

```python
position_counts = df['Position'].value_counts()

position_counts.plot(kind='bar', title='Employee Distribution by Position', color='orange')
plt.xlabel('Position')
plt.ylabel('Number of Employees')
plt.tight_layout()
plt.show()
```

```python
# Create age groups
bins = [20, 30, 40, 50, 60]
labels = ['21-30', '31-40', '41-50', '51-60']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

age_group_counts = df['AgeGroup'].value_counts().sort_index()

age_group_counts.plot(kind='bar', title='Employee Age Group Distribution', color='green')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.tight_layout()
plt.show()
```

```python
team_salary = df.groupby('Team')['Salary'].sum().sort_values(ascending=False)
position_salary = df.groupby('Position')['Salary'].sum().sort_values(ascending=False)

team_salary.plot(kind='bar', title='Salary Expenditure by Team', color='purple')
plt.xlabel('Team')
plt.ylabel('Total Salary')
plt.tight_layout()
plt.show()

position_salary.plot(kind='bar', title='Salary Expenditure by Position', color='red')
plt.xlabel('Position')
plt.ylabel('Total Salary')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Calculate correlation
correlation = df['Age'].corr(df['Salary'])
print("Correlation between Age and Salary:", correlation)

# Scatterplot
sns.scatterplot(data=df, x='Age', y='Salary')
plt.title('Age vs Salary')
plt.tight_layout()
plt.show()
```
